```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting a Vector

```python
# quiver - it is actually object used on a person's back to hold arrows
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
plt.quiver(0,0,-3,-6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of Two Vectors

```python
vec_1 = np.asarray([0,0,2,3])
vec_2 = np.asarray([0,0,3,-2])
sum = vec_1 + vec_2
print(sum)
```

```python
# vector 1
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
# vector 2
plt.quiver(0,0,3,-2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
# sum vector
plt.quiver(0,0,5,1, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')

# x and y axes limits
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.show()
```

Subraction of Two Vectors

```python
vec_1 = np.asarray([0,0,2,3])
vec_2 = np.asarray([0,0,3,-2])
diff = vec_1 - vec_2
print(diff)
```

```python
# vector 1
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'b')
# vector 2
plt.quiver(0,0,3,-2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')
# diff vector
plt.quiver(0,0,-1,5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')

# x and y axes limits
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.show()
```

Multiplying a Vector by a Scalar

```python
vec_1 = np.asarray([0,0,2,3])

vec_2 = 2 * vec_1
print(vec_2)
```

```python
# vector 1
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
# vector 2
plt.quiver(0,0,4,6, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')

# x and y axes limits
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.show()
```

```python
vec_1 = np.asarray([0,0,2,3])

vec_2 = -0.5 * vec_1
print(vec_2)
```

```python
# vector 1
plt.quiver(0,0,2,3, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r')
# vector 2
plt.quiver(0,0,-1,-1.5, scale_units = 'xy', angles = 'xy', scale = 1, color = 'g')

# x and y axes limits
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.show()
```
